```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('no_prepro_loan.csv')
df.head()
```

```python
y=df[' loan_status']

x=df.drop(' loan_status',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb_model=xgb.fit(xtrain,ytrain)
y_pred_train = xgb_model.predict(xtrain)
y_pred_test = xgb_model.predict(xtest)

print("The test accuracy score is: ",accuracy_score(ytest,y_pred_test))
```
